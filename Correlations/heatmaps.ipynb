```
import pandas as pd
import numpy as np
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data=pd.read_csv('variables.csv',index_col=0)
```

**Small correlation matrix:**

```
reduced=data.drop(['language','H_20','R_20','H_60','R_60','H_100','R_100', 'H_5000','R_5000' ], axis=1)
corr=reduced.corr('spearman')
#triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

sns.heatmap(corr, annot=False,
        xticklabels=corr.columns,
        yticklabels=corr.columns, mask=mask, cmap=cmap, linewidths=.4, cbar_kws={"shrink": .6})

ax.tick_params(labelsize=13)
```

**Big correlation matrix:**

```
corr=data.drop(['language'], axis=1).corr('spearman')
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(20, 20))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, fmt='.2f')
```
